```python

import tensorflow as tf

키 = 170
신발 = 260

a = tf.Variable(0.1)
b = tf.Variable(0.2)

def 손실함수():
    예측값 = 키 * a + b
    return tf.square(신발 - 예측값)

opt = tf.keras.optimizers.Adam(learning_rate=0.1)

for i in range(100):
    with tf.GradientTape() as tape:
        loss = 손실함수()
    grads = tape.gradient(loss, [a, b])
    opt.apply_gradients(zip(grads, [a, b]))

print("a:", a.numpy())
print("b:", b.numpy())
print("예측:", 키 * a.numpy() + b.numpy())
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('gpascore.csv')

#data.isnull().sum()
data = data.dropna() #data.fillna(100)

y데이터 = data['admit'].values

x데이터 = []
for i, rows in data.iterrows():
  x데이터.append([rows['gre'],rows['gpa'],rows['rank']])
print(x데이터)

```

```python
# 딥러닝 모델만들기

model= tf.keras.models.Sequential([
  tf.keras.layers.Dense(64, activation='tanh'), #노드의개수
  tf.keras.layers.Dense(128, activation='tanh'),
  tf.keras.layers.Dense(1,activation='sigmoid'), #마지막출력레이어는 답 하나여야하면 1임
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

#학습
model.fit(np.array(x데이터),np.array(y데이터),epochs=5)

#새로운 값으로 예측
예측값 = model.predict(np.array([[750,3.70,3], [400,2.2,1] ] ))

print(예측값)
```
